# nlargest & nsmallest

They will give us the largest and smallest vakues. For example: The 10 largest prices.

```python
houses.price.nlargest()  #by default, we get the 5 largest values.
```

Now, let´s go ahead and get the 9 largest numbers

```python
houses.price.nlargest(9)
```

Now, let´s get the 10 cheapest homes

```python
houses.price.nsmallest(10)
```

# The keep parameter in nlargest and nsmallest

**keep** will keep al accurrences. This can result in a series of size larger than n.

```python
houses.price.nlargest(7, keep="all")  #Top 7 prices
```

We want Pandas to give us all the information of the top 10 rowsranked by price

```python
houses.nlargest(10, ["price"])
```

Let´s sort by bedrooms. We want to know, which homes have the largest bedrooms.

```python
houses.nlargest(10,["bedrooms"]) 
```

```python
houses.nlargest(10,["bedrooms", "bathrooms"]) #we can also include + columns. First, we soft by bedrooms and then by bathrooms.
```

# The power of value_counts() method

This is one of the most important methods in series. According to the Pandas documentation...
The method return a series containing counts of unique values.

```python
houses["bedrooms"] #This is a series
```

```python
houses["bedrooms"].value_counts() #it tells me how many times the value occuers. Ex: 3 bedrooms occuer 9824 times.
```

This gets our data in a shape that is very easy to plot.
We often want to group things or count things by value.

```python
houses["floors"].value_counts()  #analyzing the number of floors with x amount of flooers.
```
